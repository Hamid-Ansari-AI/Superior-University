### Program for managing different types of documents:

```python
import csv

class Document:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Author: {self.author} \nTitle: {self.title}")
        
class  Book(Document):
    def __init__(self, title, author, genre=None, pages=None):
        super().__init__(title, author)
        self.genre = genre
        self.pages = pages
    
    def display_info(self):
        super().display_info()
        if self.genre:
            print(f"Genre: {self.genre}")
        if self.pages:
            print(f"Pages: {self.pages}")

class Article(Document):
    def __init__(self, title, author, journal=None, doi=None):
        super().__init__(title, author)
        self.journal = journal
        self.doi = doi

    def display_info(self):
        super().display_info()
        if self.journal:
            print(f"Journal: {self.journal}")
        if self.doi:
            print(f"DOI: {self.doi}")

def store_info(document):
    with open("Documents.csv", "a", newline="") as file:
        writer = csv.writer(file)
        if file.tell() == 0:
            writer.writerow(["Title", "Author", "Genre", "Pages", "Journal", "DOI"])

        if isinstance(document, Book):
            writer.writerow([document.title, document.author, document.genre if document.genre else "", document.pages if document.pages else "", "", ""])
        elif isinstance(document, Article):
            writer.writerow([document.title, document.author, "", "", document.journal if document.journal else "", document.doi if document.doi else ""])

    print(f"Information successfully stored.\n")


def retrieve_info():
    try:
        with open("Documents.csv", "r", newline="") as file:
            reader = csv.DictReader(file)
            count = 1
            for row in reader:
                output = f"{count}. {row['Title']}, {row['Author']}"
                if row['Genre']:
                    output += f", {row['Genre']}"
                if row['Pages']:
                    output += f", {row['Pages']}"
                if row['Journal']:
                    output += f", {row['Journal']}"
                if row['DOI']:
                    output += f", {row['DOI']}"
                print(output)
                count += 1

    except FileNotFoundError:
        print("No information found.")
```

```python
book1 = Book("Python Programming", "Hamid", "Technology", 350)
book2 = Book("c++", "Arslan", "Technology")
book3 = Book("Java", "Ahad", pages=450)
book4 = Book("Trust & Inpire", "Stephen Covey")

# Displaying book info
book1.display_info()
store_info(book1) # Store in csv

book2.display_info()
store_info(book2) # Store in csv

book3.display_info()
store_info(book3) # Store in csv

book4.display_info()
store_info(book4) # Store in csv

# Creating Article objects
article1 = Article("AI in Healthcare", "Afzal", "Medical Journal", "10.1/ai.medical")
article2 = Article("AI in Graphics", "Fahad", "Medical Journal")
article3 = Article("Cyber Security", "Saad", doi="20.1/cs.technology")
article4 = Article("Quantum Computing", "Usman") 

# Displaying article info
article1.display_info()
store_info(article1) # Store in csv

article2.display_info()
store_info(article2) # Store in csv

article3.display_info()
store_info(article3) # Store in csv

article4.display_info()
store_info(article4) # Store in csv

# Retrieve from CSV
retrieve_info()
```
